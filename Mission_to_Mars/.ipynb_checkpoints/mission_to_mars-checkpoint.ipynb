```python
# Dependencies
import pymongo
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from flask import Flask, render_template, redirect
#from flask_pymongo import PyMongo
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

## NASA Mars News

```python
# URL of page to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# Retrieving and assigning latest news title
news_title = soup.body.find_all('div', class_='content_title')[0].text

news_title
```

```python
# Retrieving and assigning latest news paragraph

news_p = soup.body.find_all('div', class_='article_teaser_body')[0].text

news_p
```
